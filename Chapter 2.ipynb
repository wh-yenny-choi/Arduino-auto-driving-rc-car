# Chapter2 아두이노 스마트 RC카 시작하기
1. 아두이노 IDE 설치하기
2. 아두이노 IDE 이해하기
3. 아두이노 IDE 살펴보기

## 01. 아두이노 IED 설치하기
자율주행 RC카 sw구현, 컴파일, 업로드 위한 개발 환경 아두이노 스케치 IDE 설치
- 통합개발환경(IDE) = 스케치 작업 + 컴파일 + 업로드



## 02. 아두이노 IED 이해하기
1. 툴바
- 컴파일 : 소스를 제대로 작성했는지 확인
- 업로드 : 컴파일 완료시, 아두이노 보드에 업로드
- 새파일 : 새 스케치 문서 탭
- 열기 : 기존 작성 스케치 문서 불러오기
- 저장 : 현재 작업중 스케치 문서 저장
- 시리얼 모니터 : 아두이노 보드로부터 메시지를 확잉ㄴ하거나 아두이노 보드에게 메시지를 보낼 수 있음
2. 탭목록 
3. 텍스트 에디터 : 프로그램 코드 작성 부분
4. 콘솔 : 컴파일 및 업로드 시 정상 및 에러 유무에 대해 표시되는 부분

## 03. 아두이노 IED 살펴보기
- 명령어는 {로 시작해서}로 끝을 맺는다
- 스케치는 크게 setup()과 loop() 두 개의 중요한 메서드로 구성
    - setup() 함수 : 아두이노가 동작 시작시 필요한 작업 작성 부분. 핀번호 등 실행 시 필요한 것들에 설정해주는 영역
    - loop() 함수 : 아두이노가 setup실팽 후 계속해서 반복적으로 해야 할 작업들을 작성하는 부분. 실제적으로 아두이노 보드가 수행해야 할 일들을 프로그래밍해 주는 영역
- 스케치 순서는 일반적으로 프로그래밍 → 컴파일 → 업로딩 순서로 진행

스케치 프로그램의 라이프 사이클

Arduino Start → Setup() → loop() → Arduino reset
